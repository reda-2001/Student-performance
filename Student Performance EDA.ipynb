<h3 style="color:blue">Interpretations</h3><br>
<b>
    <ul>
        <li>We can conclude that mean = median = mode approximativelly  </li>
        <li>median = mean . The data is quite symmetrical(not skewed).In this case, the mean is a good representation of the               central tendency,because the mean is sensitive to outliers.<font color="red">(In general, the mean is considered to be         a good measure of central tendency for symmetrical data, while the median is considered to be a better measure for               skewed data.)</font></li>
        <li>the mean is : 66.09 , the range is 100 - 0 = 100 . This suggest that the majority of student got a mark around 66           with some that got exellent marks and some that got bad marks</li>
        <li>The range is 100 and the interquantile range is : 77-57=20. In this case, the range is significantly           larger than the IQR. This suggests that there is at least one value in thedataset that is much higher or lower than the         majority of the other values, which is causing the range to be larger than the IQR.It may also indicate that the data           are more dispersed and there is a greater degree of variability in the dataset.</li>
